# Chatbot with Regex, Intent Matching, and Entities: AlienBot Exercise

```python
# importing regex and random libraries
import re
import random

class AlienBot:
  # potential negative responses
  negative_responses = ("no", "per niente", "zero", "macché", "mi dispiace", "non posso proprio", "ma non rompere")
  # keywords for exiting the conversation
  exit_commands = ("esci", "pausa", "fine", "arrivederci", "ciao", "vado")
  # random starter questions
  random_questions = (
        "Vuoi bere il mio muco?\n> ",
        "Ma gli altri membri della tua specie sono tutti così attraenti?\n> ",
        "Anche a te piace mangiare la spazzatura?\n> ",
        "Ci sono roditori intelligenti su questo pianeta?\n> ",
        "Chi è il capo del tuo pianeta?\n> ",
        "Su quali altri pianeti sei stato questa settimana?\n> ",
        "Qual è la specie che produce più formaggio su questo pianeta?\n> "
    )

  def __init__(self):
    self.alienbabble = {'answer_why_intent': [r'.*perch.*?'], 'describe_planet_intent': [r'.*tuo.*pianet.*'], 'cubed_intent': [r'^.*cubo di (\d+).*', r'^(\d+) al cubo.*', r'.*\s(\d+) al cubo.*'], 'favorite_food_intent': [r'.*cibo.*', r'.*mangiare.*']}

  # Define .greet() below:
  def greet(self):
    self.name = input("Salve, sono AstroViscidino. Tu come ti chiami?\n> ")
    will_help = input(f"Piacere di conoscerti, {self.name}! \nÈ il mio primo giorno sulla terra. \nVuoi aiutarmi a conoscere questo pianeta?\n> ").lower()
    for negative_response in self.negative_responses:
      if negative_response in will_help:
        print("D'accordo, tornerò a spremermi le mie ghiandoline segrete! \nHihihi! \nCiao ciao.")
        return
    self.chat()

  # Define .make_exit() here:
  def make_exit(self, reply):
    for exit_command in self.exit_commands:
      if exit_command in reply:
        print("Va bene, torno a palparmi le sopracciglia posteriori!\nHihihi!\nCiao ciao!")
        return True

  # Define .chat() next:
  def chat(self):
    reply = input(random.choice(self.random_questions)).lower()
    while not self.make_exit(reply):
      reply = input(self.match_reply(reply))
      # reply = input("Vuoi tastare i miei pori lubrificanti?\n> ")

  # Define .match_reply() below:
  def match_reply(self, reply):
    for key, value in self.alienbabble.items():
      for regex in value:
        found_match = re.match(regex, reply.lower())
        if found_match and key == 'describe_planet_intent':
          return self.describe_planet_intent()
        elif found_match and key == 'answer_why_intent':
          return self.answer_why_intent()
        elif found_match and key == 'favorite_food_intent':
          return self.favorite_food_intent()
        elif found_match and key == 'cubed_intent':
          return self.cubed_intent(found_match.groups()[0])
    return self.no_match_intent()

  # Define .describe_planet_intent():
  def describe_planet_intent(self):
    responses = ("Provengo da un'enorme distesa di verruche fumanti.\n> ", "Sul mio pianeta mangiamo solo broccoli lessi.\n> ", "Il mio pianeta è pieno di puzze succulente.\n> ")
    return random.choice(responses)

  # Define .answer_why_intent():
  def answer_why_intent(self):
    responses = ("Perché ne avevo voglia! Hihihi!\n> ", "Voi umani siete creature curiose! \nAdesso fruga dentro il mio ombelico.\n> ", "Non ci sono spiegazioni. \nHo solo tanta voglia di conoscere nuove puzze.\n> ")
    return random.choice(responses)
       
  # Define .cubed_intent():
  def cubed_intent(self, number):
    cube = int(number) ** 3
    return f"Il cubo di {number} è {cube}!\n> "

  # Define .no_match_intent():
  def no_match_intent(self):
    responses = ("Oh, wow! \nMi fai un'altra domanda?\n> ", "Scusa, ho le orecchie piene di muffa. \nPuoi farmi un'altra domanda?\n> ", "Divertente! \nMa sai che io so dirti quanto fa un numero al cubo?\n> ", "Bello! \nVuoi farmi una domanda riguardo al mio pianeta?\n> ", "Sei proprio interessante. \nMi vuoi chiedere perché sono qui?\n> ", "Vorrei davvero dirtelo. \nChiedimi di più.\n> ", "Mi sto innamorando di te. Parlami ancora.\n> ", "Sento puzza di domandina interessante! \nProva di nuovo.\n> ")
    return random.choice(responses)

    #Define .favorite_food_intent():
  def favorite_food_intent(self):
    responses = ("Mi piace mangiare le croste.\n> ", "Un bel tortino di pustole.\n> ", "Tutto quello che secernono i miei pori.\n> ", "Di tutto! Basta che faccia una bella puzza succosa.\n> ", "Uova strapazzate condite con unghie appena tagliate!\nNon riesco a dire di no.\n> ", "Beh, amo il gorgonzola di troll.\n> ")
    return random.choice(responses)

# Create an instance of AlienBot below:
inizio_alienbotto = AlienBot()
inizio_alienbotto.greet()
```
