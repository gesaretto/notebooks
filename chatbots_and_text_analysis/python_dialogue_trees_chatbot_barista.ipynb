```python
# Define your functions
total_order = []

## Coffee Bot Function
def coffee_bot():
  print("Welcome to the cafe!")
  get_order()
  name = input("Can I get your name please? \n> ")
  print("Thanks, " + name + "! Your order will be ready shortly.")
  print("You have ordered:")
  for i in total_order:
    print(i)
  print("Coming up!")

## Ask if more needs are wanted
def more_drinks():
  res = input("Would you like another drink? \n[a] Yes. \n[b] No, thanks. \n> ")
  if res == "a": 
    get_order()
  elif res == "b":
    print("That's it, then.")
  else:
    print_message()
    res = more_drinks()
    return res 

## Get your order
def get_order():
  size = get_size()
  drink = get_drink_type()
  temperature = hot_iced()
  vessel = kind_cup()
  order = temperature + " " + size + " " + drink + " in a " + vessel
  total_order.append(order)
  print("Alright, that's a " + order + "!")
  more_drinks()

## Get the size of the coffee
def get_size():
  res = input("What size drink can I get for you? \n[a] Small \n[b] Medium \n[c] Large \n> ")
  if res == "a":
    return "small"
  elif res == "b":
    return "medium"
  elif res == "c":
    return "large"
  else:
    print_message()
    res = get_size()
    return res

# Get the type of drink
def get_drink_type():
  res = input("What type of drink would you like? \n[a] Brewed Coffee \n[b] Mocha \n[c] Latte \n> ")
  if res == "a":
    return "brewed coffee"
  elif res == "b":
    return "mocha"
  elif res == "c":
    return order_latte()
  else:
    print_message()
    res = get_drink_type()
    return res

# Get the kind of milk for the latte
def order_latte():
  res = input("And what kind of milk for your latte? \n[a] 2% milk \n[b] Non-fat milk \n[c] Soy milk \n> ")
  if res == "a":
    return "latte"
  elif res == "b":
    return "non-fat latte"
  elif res == "c":
    return "soy latte"
  else:
    print_message()
    res = order_latte()
    return res

# Get the kind of cup
def kind_cup():
  res = input("Do you need a plastic cup for your drink? Or do you have a reusable cup? \n[a] A plastic cup. \n[b] I have my own reusable cup. \n> ")
  if res == "a":
    print("Geez, you suck dude.")
    return "plastic cup"
  if res == "b":
    return "reusable cup"
  else:
    print_message()
    res = kind_cup()
    return res

# Get the drink hot or iced
def hot_iced():
  res = input("Would you like it hot or iced? \n[a] Hot. \n[b] Iced. \n> ")
  if res == "a":
    return "hot"
  elif res == "b":
    return "iced"
  else:
    print_message()
    res = hot_iced()
    return res

# A polite message
def print_message():
  print("I'm sorry. I did not understand your selection. Please enter the corresponding letter for your response.")

# Call coffee_bot()!
coffee_bot()
```
