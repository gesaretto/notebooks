```python
import pandas as pd
from matplotlib import pyplot as plt

healthcare = pd.read_csv("healthcare.csv")
healthcare.columns = healthcare.columns.str.strip()
# print(healthcare.columns)
# print(healthcare.head())

print(healthcare["DRG Definition"].unique())
# "Chest pain" is '313- CHEST PAIN.'


def show_boxplots_by_state(column_name, diagnosis_name):
  plt.figure(figsize=(20, 6))
  exp_per_diag = healthcare[healthcare["DRG Definition"] == diagnosis_name]
  plt.boxplot([exp_per_diag[exp_per_diag["Provider State"] == state][column_name].values for state in exp_per_diag["Provider State"].unique().tolist()], labels = exp_per_diag["Provider State"].unique().tolist())
  plt.title(diagnosis_name + " - " + column_name)
  plt.show()

show_boxplots_by_state("Average Covered Charges", "313 - CHEST PAIN")
show_boxplots_by_state("Average Medicare Payments", "917 - POISONING & TOXIC EFFECTS OF DRUGS W MCC")
```
