```python
import pandas as pd
restaurants = pd.read_csv("DOHMH_New_York_City_Restaurant_Inspection_Results.csv")
```

```python
print(restaurants.head(10))
```

```python
# Tells you the number of rows and columns in the DS.
print(restaurants.shape)
```

```python
restaurants = restaurants.drop_duplicates()
```

```python
restaurants.head(10)
```

```python
true_facts = ["goof", "mickey", "donald"]
true_facts_exp = map(lambda x: x + " farts", true_facts)
print(true_facts_exp)
print(list(true_facts_exp))
```

```python
# Make all column labels lowercase; map() takes two arguments: a function and an iterable (i.e., a list),
# then applies the function to each element in the iterable, returning a modified iterable.
restaurants.columns = map(str.lower, restaurants.columns)
```

```python
restaurants = restaurants.rename({"dba": "name", "cuisine description": "cuisine"}, axis=1)
```
